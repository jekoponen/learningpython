```python
from geopy.geocoders import Nominatim
nom=Nominatim(scheme="http")     #create Nominatim object called nom
```

```python
import pandas
df=pandas.read_csv("supermarkets.csv")
```

```python
df["Concatenated"]=df["Address"]+","+df["City"]+", ,"+df["State"]+","+df["Country"]+"," #concatenate full address to new column 
# Nominatim.geocode method takes comma separated full address as argument
```

```python
df
```

```python
df["Coordinates"]=df["Concatenated"].apply(nom.geocode) #.apply is panda method which applies .geocode method to all rows in dataFrame.
```

```python
df # Coordinates has now location object for all existing addresses
```

```python
df["Latitude"]=df["Coordinates"].apply(lambda x: x.latitude) #x is temporary variable. this give error cos there is row with NoneType object (coordinates wasn't found for address)
```

```python
df["Latitude"]=df["Coordinates"].apply(lambda x: x.latitude if x != None else None) #Add condition, apply .latitude method to x if x NOT None
df
```

```python
df["Longitude"]=df["Coordinates"].apply(lambda x: x.longitude if x != None else None) #same for the longitude
df
```
